```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt


# Carica il CSV originale
df_originale = pd.read_csv('./filtered_delivery_data.csv', sep='\t')

# Supponiamo che la colonna si chiami 'SHIPPING_DATE'
# Estrai solo il mese (6° e 7° carattere)
df_originale['SHIPPING_DATE'] = df_originale['SHIPPING_DATE'].str[5:7]

# Stampa il risultato per verificare
print(df_originale['SHIPPING_DATE'])

# Conta quante volte appare ogni mese
conteggio_mesi = df_originale['SHIPPING_DATE'].value_counts()

# Stampa il conteggio dei mesi
print(conteggio_mesi)
```

```python
import matplotlib.pyplot as plt

# Conta quante volte appare ogni tipo di SERVICETYPE
conteggio_service_type = df_originale['SERVICETYPE'].value_counts()

# Crea il grafico a torta
plt.figure(figsize=(8, 6))
plt.pie(conteggio_service_type, labels=conteggio_service_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuzione dei tipi di SERVICETYPE')
plt.axis('equal')  # Per mantenere la forma circolare
plt.show()
```

```python
# Conta quante volte appare ogni codice postale nella colonna 'DEPARTURE_ZIPCODE'
conteggio_departure_zipcode = df_originale['DEPARTURE_ZIPCODE'].value_counts()

# Calcola la percentuale per ogni codice postale
percentuale_zipcode = conteggio_departure_zipcode / conteggio_departure_zipcode.sum() * 100

# Filtra i codici postali con percentuale maggiore del 2%
zipcodes_frequenti = percentuale_zipcode[percentuale_zipcode > 2]

# Raggruppa tutti gli altri codici postali nella categoria "Altro"
altro = percentuale_zipcode[percentuale_zipcode <= 2].sum()
zipcodes_frequenti['Altro'] = altro

# Crea il grafico a torta
plt.figure(figsize=(8, 6))
plt.pie(zipcodes_frequenti, labels=zipcodes_frequenti.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuzione dei codici postali di partenza (DEPARTURE_ZIPCODE) - Maggiore del 2%')
plt.axis('equal')  # Per mantenere la forma circolare
plt.show()
```
